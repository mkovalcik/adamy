## Pearson Coefficient heatmap for three sectors

First let's re-draw the heatmap of pearson coefficient: which is great in revealing the linear relationship between two features. This time, let's put EV/EBITA in.

Recap on categories of these features:

- size: `t_ev`, `t_rev`, `ebitda`
- profitability: `ebitda_margin`, `return_on_assets`, `return_on_equity`, `ebitda/interest_exp`
- growth: `est_ann_rev_gr_minus_1yr_minus_capiq`, `est_ann_ebitda_gr_minus_1yr_minus_capiq`, `t_rev_1yr_growth`, `ebitda_1yr_growth`, `t_rev_3_yr_cagr`, `ebitda_3yr_cagr`, `t_rev_5yr_cagr`, `ebitda_5yr_cagr`, `capex_as_percent_rev`
- risk: `t_debt/cap_percent`, `t_debt/equity_percent`

```python
col = ['t_ev/ebitda', 't_ev', 
       't_rev', 'ebitda',
       'ebitda_margin', 'return_on_assets', 'return_on_equity', 
       'est_ann_rev_gr_1yr', 'est_ann_ebitda_gr_1yr', 
       't_rev_1yr_growth', 'ebitda_1yr_growth', 't_rev_3_yr_cagr', 'ebitda_3yr_cagr', 
       't_rev_5yr_cagr', 'ebitda_5yr_cagr', 'capex_as_percent_rev', 
       't_debt/cap_percent', 't_debt/equity_percent', 'ebitda/interest_exp']
```

### consumerDiscrete

```python
sns.set(font_scale=2)
plt.figure( figsize=(16,16) )
plt.title( 'Pearson Correlation of Features', y=1.05, size=20)
fig = sns.heatmap( raw_data_frame[0][col].corr(), linewidths=0.1, vmax=1.0, 
            square=True, linecolor='white', annot=True,annot_kws={"size": 10} )
```
